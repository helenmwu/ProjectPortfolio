```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'http://books.toscrape.com/catalogue/category/books/romance_8/index.html'

page = requests.get(url)

soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
table = soup.find_all(class_ = 'image_container')
```

```python
print(table)
```

```python
book_titles = []
```

```python
for i in table:
    text = i.find_all('img', alt = True)
    book_titles.append(text)
 
```

```python
print(book_titles)
```

```python
import re
```

```python
title_only = []

for title in book_titles:
    title = str(title)
    t = re.findall(r'"(.*?)(?<!\\)"', title)
    title_only.append(t)
    
```

```python
final = []
for i in title_only:
    for j in i:
        final.append(j)
        break
        
```

```python
print(final)
```
